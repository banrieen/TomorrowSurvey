```python
# 饼图
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt

from io import BytesIO
from lxml import etree
import base64



lable =('A','B','C','D')
fras= [15,30,45,10]
lable =('pass', 'fail')
fras=[52, 1]
explode=[0,0.2]#设置离员中心的位置 此处是为了突出显示B （不要此参数，就是一个员）
plt.axes(aspect=1) #此处设置的目的 是为了让饼状图画出来是圆形
plt.pie(x=fras,labels=lable,autopct='%.2f%%',explode=explode,shadow=True) #传入数据及参数，占比保留两位小数 explode突出显示 shadow 阴影
# plt.subplots_adjust(wspace = 0)


# figure 保存为二进制文件
buffer = BytesIO()
plt.savefig(buffer)  
plot_data = buffer.getvalue()
imb = base64.b64encode(plot_data)  
#imb = plot_data.encode('base64')   # 对于 Python 2.7可用 
iris_des = """<h1>Iris Describe Stastic</h1>"""
ims = imb.decode()
imd = "data:image/png;base64,"+ims

iris_im = """<h1>Iris Figure</h1>  """ + """<img src="%s">""" % imd   
 
root = "<title>Iris Dataset</title>"
root = root + iris_des + iris_im  #将多个 html 格式的字符串连接起来
 
# lxml 库的 etree 解析字符串为 html 代码，并写入文件
html = etree.HTML(root)
tree = etree.ElementTree(html)
tree.write('iris.html')


# # 图像数据转化为 HTML 格式
# imb = base64.b64encode(plot_data) 

plt.show()
```
